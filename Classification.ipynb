```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
data=pd.read_csv("dataset.csv")
data["Class"]=(data["Class"]=="Malicious").astype(int)
```

```python
data = data.dropna(how='any',axis=0)
```

```python
sns.heatmap(data.corr(),annot=True)
```

```python
x=data[["pdfsize","metadata size", "pages","xref Length","title characters","isEncrypted","embedded files","stream","trailer","encrypt","ObjStm","Colors"]]
y=data["Class"]
```

```python
data.isnull().sum()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
print(model.score(x_test,y_test),model.score(x_train,y_train))
```

```python
tree=XGBClassifier()
tree.fit(x_train,y_train)
```

```python
print(tree.score(x_test,y_test),tree.score(x_train,y_train))
```
